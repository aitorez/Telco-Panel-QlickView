```python
import random
import csv
import string
import os
from datetime import datetime, timedelta

# Define the header for the CSV file
header = ['typeservice', 'namecdr', 'b_numerollamante', 'tarifa', 'b_numerollamado', 'fecha', 'hora', '?', 'y', 'bytes', 'tecnolog√≠a', 'importe']

# Define the number of files you want to create
num_files = 10

# Define the number of rows you want in each file
num_rows = 100

# Define the directory where you want to save the files
dir_path = "C:/Users/aitor/OneDrive/Escritorio/Anaconda/boot-camp-QlickView/random CDRs"

# Create the directory if it doesn't exist
if not os.path.exists(dir_path):
    os.makedirs(dir_path)

# Generate and save the CSV files
for i in range(num_files):
   # Generate a random filename for the CSV file
    rand_num = str(random.randint(10000000, 99999999))
    date_str = datetime.now().strftime("%d_%m_%Y")
    filename = "vod_" + rand_num + "_" + date_str + ".csv"
    file_path = os.path.join(dir_path, filename)

    # Create a new CSV file and write the header row
    with open(file_path, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(header)

        # Write the data rows
        for j in range(num_rows):
            # Generate a random date within the range of 30 days from today
            date = datetime.now() - timedelta(days=random.randint(0, 30))
            # Generate a random value for the "bytes" column between 1 MB and 10 GB
            bytes_val = random.randint(1, 1000) * 1024 * 1024
            # Generate a random value for the "importe" column between 0 and 100
            importe_val = round(random.uniform(0, 100), 4)

            row = [
                'DAT',
                ''.join(random.choices(string.ascii_uppercase + string.digits, k=10)),
                random.randint(100000000, 999999999),
                '10 GB y llamadas ilimitadas',
                '',
                date.strftime("%d-%m-%Y"),
                '00:00:00',
                '0',
                '0',
                bytes_val,
                'Datos 4G',
                importe_val
            ]
            writer.writerow(row)
```
